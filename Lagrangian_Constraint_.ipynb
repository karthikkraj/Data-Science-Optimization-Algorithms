# **Objective Function:**

This code defines the objective function that the optimization algorithm aims to minimize.

**Functionality:**

* Mathematical Representation: The objective function is defined as: (x1 - 2)^2 + (x2 - 1)^2.

* Input: It takes a 2D point x as input, represented as an array-like object (e.g., a NumPy array or a list). x[0] represents the x1-coordinate, and x[1] represents the x2-coordinate of the point.

* Calculation: The function calculates the value of the objective function at the given point x using the formula mentioned above.

* Output: It returns the calculated value of the objective function at the point x.

* Global Minimum: The function has a global minimum at the point x = [2, 1], where the objective function value is 0.

**Purpose:**

* This function is crucial in the optimization process as it defines the quantity that needs to be minimized. The optimization algorithm will iteratively search for the values of x1 and x2 that result in the lowest possible value of this objective function, while also satisfying any constraints.

```python
# Objective Function
def objective_function(x):
    return (x[0] - 2)**2 + (x[1] - 1)**2
```

## **Gradient of the Objective Function:**

This function calculates the gradient of the objective function. The gradient is a vector that points in the direction of the steepest ascent of the function. In optimization, it's used to find the direction of steepest descent, which helps the algorithm move towards the minimum.

## **Functionality:**

* Input: The function takes a 2D point x as input, represented as an array-like object (e.g., a NumPy array or a list).
* Calculation: It calculates the gradient of the objective function (x1 - 2)^2 + (x2 - 1)^2. The gradient is a vector with two components:
 * The partial derivative with respect to x1: 2*(x[0] - 2)
 * The partial derivative with respect to x2: 2*(x[1] - 1)
* Output: The function returns the gradient vector as a NumPy array [2*(x[0] - 2), 2*(x[1] - 1)].

## **Purpose:**

* In optimization algorithms like gradient descent, the gradient is used to update the current solution and move towards a minimum. By taking steps in the opposite direction of the gradient (steepest descent), the algorithm iteratively converges towards a point where the objective function is minimized

```python
# Gradient of the Objective Function
def objective_gradient(x):
    return np.array([2*(x[0] - 2), 2*(x[1] - 1)])
```

## **Equality Constraint:**
This function defines the equality constraint for the optimization problem. The equality constraint represents a circle with radius 1 centered at the origin (0, 0).

## **Functionality:**

* Input: It takes a 2D point x as input, represented as an array-like object (e.g., a NumPy array or a list). x[0] represents the x1-coordinate, and x[1] represents the x2-coordinate of the point.

* Calculation: The function calculates the value of the equality constraint equation: x1^2 + x2^2 - 1.

* Output: It returns the calculated value of the constraint equation.
 * If the point x lies on the circle (satisfies the constraint), the function will return 0.
 * If the point x is not on the circle, the function will return a non-zero value, indicating the degree of constraint violation.

## **Purpose:**

* This function is used to ensure that the solution found by the optimization algorithm satisfies the equality constraint. The augmented Lagrangian method incorporates this constraint into the objective function, penalizing solutions that violate the constraint.

```python
# Equality Constraint
def equality_constraint(x):
    return x[0]**2 + x[1]**2 - 1
```

## **Gradient of the Equality Constraint:**

This function calculates the gradient of the equality constraint. The gradient is a vector that points in the direction of the steepest ascent of the constraint function. In constrained optimization, it's used to incorporate the constraint into the optimization process.

## **Functionality:**

* Input: It takes a 2D point x as input, represented as an array-like object (e.g., a NumPy array or a list).

* Calculation: It calculates the gradient of the equality constraint function x1^2 + x2^2 - 1. The gradient is a vector with two components:

 * The partial derivative with respect to x1: 2*x[0]
 * The partial derivative with respect to x2: 2*x[1]

* Output: The function returns the gradient vector as a NumPy array [2*x[0], 2*x[1]].

## **Purpose:**

* The gradient of the equality constraint is used in the augmented Lagrangian method to update the Lagrange multiplier associated with the constraint.

* It helps to adjust the optimization process to ensure that the solution satisfies the equality constraint.

* The gradient provides information about the direction in which the constraint function changes most rapidly, allowing the algorithm to move towards satisfying the constraint.

```python
# Gradient of the Equality Constraint
def equality_constraint_gradient(x):
    return np.array([2*x[0], 2*x[1]])
```

## **Gradient of the Inequality Constraint:**

This function calculates the gradient of the inequality constraint. Similar to the equality constraint, the gradient here represents the direction of the steepest ascent of the constraint function. This information is crucial for incorporating the inequality constraint into the optimization process.

## **Functionality:**

* Input: The function takes a 2D point x as input, represented as an array-like object. However, in this specific case, the gradient is constant and does not depend on the input x.

* Calculation: The inequality constraint is x1 + x2 ≥ 0, which can be rewritten as -x1 - x2 ≤ 0 to conform to the standard form of inequality constraint in optimization problem. The gradient of this constraint is a constant vector [-1, -1]. This means the direction of steepest ascent for the inequality constraint function is always the same, regardless of the point x.

* Output: The function returns the gradient vector as a NumPy array [-1, -1].

## **Purpose:**

* In the augmented Lagrangian method, the gradient of the inequality constraint is used to update the Lagrange multiplier associated with the inequality constraint.

* It ensures that the optimization algorithm considers the inequality constraint and tries to find a solution that satisfies it.

* By incorporating this gradient information, the algorithm adjusts its search direction to avoid violating the inequality constraint.

```python
# Gradient of the Inequality Constraint
def inequality_constraint_gradient(x):
    return np.array([-1, -1])
```

# **Gradient of the Augmented Lagrangian**

This function calculates the gradient of the augmented Lagrangian function, which is a crucial component of the augmented Lagrangian method for constrained optimization. The augmented Lagrangian combines the original objective function with penalty terms for constraint violations, and its gradient guides the optimization process.

## **Functionality:**

1. Input:

 * x: The current point in the search space (a 2D point).
 * lambda_eq: Lagrange multiplier for the equality constraint.
 * lambda_ineq: Lagrange multiplier for the inequality constraint.
 * rho_eq: Penalty parameter for the equality constraint.
 * rho_ineq: Penalty parameter for the inequality constraint.

2. Calculations:
 * grad_obj: Calculates the gradient of the original objective function at point x using the objective_gradient function.
 * eq_value: Calculates the value of the equality constraint at point x using the equality_constraint function.
 * grad_eq: Calculates the gradient of the equality constraint at point x using the equality_constraint_gradient function.
 * ineq_value: Calculates the value of the inequality constraint at point x using the inequality_constraint function defined before.
 * grad_ineq: Calculates the gradient of the inequality constraint at point x using the inequality_constraint_gradient function.
 * eq_term: Calculates the penalty term for the equality constraint using the Lagrange multiplier, penalty parameter, and gradient of the equality constraint.
 * ineq_term: Calculates the penalty term for the inequality constraint, but only if the inequality constraint is violated (ineq_value > 0). It uses the Lagrange multiplier, penalty parameter, and gradient of the inequality constraint.
3. Output:
 * Returns the gradient of the augmented Lagrangian function, which is the sum of the gradient of the objective function and the penalty terms for both equality and inequality constraints. This gradient is used to update the solution in the optimization process.

## **Purpose:**

* The augmented_lagrangian_gradient function is a core part of the augmented Lagrangian method.
* It provides the direction for updating the solution in each iteration of the optimization process.
* By incorporating the gradients of the objective function and constraints, along with penalty terms, it guides the algorithm towards a solution that minimizes the objective function while satisfying the constraints.


```python
# Gradient of the Augmented Lagrangian
def augmented_lagrangian_gradient(x, lambda_eq, lambda_ineq, rho_eq, rho_ineq):
    grad_obj = objective_gradient(x)
    eq_value = equality_constraint(x)
    grad_eq = equality_constraint_gradient(x)

    ineq_value = inequality_constraint(x)
    grad_ineq = inequality_constraint_gradient(x)

    # Penalty terms for the constraints
    eq_term = (lambda_eq + rho_eq * eq_value) * grad_eq
    ineq_term = (lambda_ineq + rho_ineq * ineq_value) * grad_ineq if ineq_value > 0 else 0

    return grad_obj + eq_term + ineq_term
```

## **Gradient Descent on the Augmented Lagrangian**


This function performs gradient descent on the augmented Lagrangian function to find a local minimum. It's an inner loop within the augmented Lagrangian method, responsible for iteratively updating the solution based on the gradient of the augmented Lagrangian.

## **Functionality:**

1. Initialization:

* positions: A list to store the history of points visited during the optimization process. It starts with the initial point x.
* step_size: The learning rate for gradient descent, controlling the size of steps taken in the direction of the negative gradient.

2. Iteration:

* The code iterates for a maximum of max_iter times.

 * In each iteration:
   * grad: Calculates the gradient of the augmented Lagrangian at the current point x using the augmented_lagrangian_gradient function.
   * Stopping Condition: If the norm (magnitude) of the gradient is less than epsilon, it means the algorithm has converged to a local minimum, and the loop breaks.
   * Update: If the stopping condition is not met, the current point x is updated by taking a step in the direction of the negative gradient: x = x - step_size * grad.
   * positions: The updated point x is appended to the positions list to track the optimization path.

3. Return:

* The function returns the final optimized solution x and the positions array containing the history of points visited during the optimization.

## **Purpose:**

* This function is a key part of the augmented Lagrangian method.
* It uses gradient descent to iteratively refine the solution, aiming to minimize the augmented Lagrangian function.
* By repeatedly updating the solution based on the gradient, it helps the algorithm converge towards a solution that satisfies the constraints and minimizes the objective function.

```python
# Gradient Descent on the Augmented Lagrangian
def minimize_augmented_lagrangian(x, lambda_eq, lambda_ineq, rho_eq, rho_ineq, max_iter=100, epsilon=1e-6):
    positions = [x.copy()]
    step_size = 0.1  # Learning rate for gradient descent

    for _ in range(max_iter):
        grad = augmented_lagrangian_gradient(x, lambda_eq, lambda_ineq, rho_eq, rho_ineq)
        if np.linalg.norm(grad) < epsilon:  # Stopping condition if gradient is small
            break
        x = x - step_size * grad
        positions.append(x.copy())

    return x, np.array(positions)
```

## **Augmented Lagrangian Method**

This function implements the augmented Lagrangian method, a technique for solving constrained optimization problems. It iteratively minimizes the augmented Lagrangian function, updates Lagrange multipliers and penalty parameters, and tracks the optimization path.

## **Functionality:**

1. Initialization:

  * x: Starts with the initial point initial_x.
  * lambda_eq, lambda_ineq: Lagrange multipliers for equality and inequality constraints, initialized to 0.
  * rho_eq, rho_ineq: Penalty parameters for equality and inequality constraints, initialized to 1.
  * all_positions: A list to store the history of points visited during optimization, starting with the initial point.

2. Outer Iteration:

  * The code iterates for a maximum of max_outer_iter times.
  * In each outer iteration:
    * Minimization: Calls minimize_augmented_lagrangian to perform gradient descent and minimize the augmented Lagrangian, updating x and appending the optimization path to all_positions.
    * Constraint Violation: Calculates the violations of equality and inequality constraints using equality_constraint and inequality_constraint.
    * Status: Prints the current iteration number, solution x, and constraint violations.
    * Convergence Check: If both constraint violations are below a threshold (1e-4), the algorithm is considered converged, and the loop breaks.
    * Update Multipliers and Penalties: If the constraint violations are  significant, the Lagrange multipliers and penalty parameters are updated to increase the penalty for constraint violations and guide the solution towards feasibility.

3. Return:

  * The function returns the all_positions array, containing the history of points visited during the optimization process.

## **Purpose:**

  * This function is the main driver of the augmented Lagrangian method.
  * It combines gradient descent with updates to Lagrange multipliers and penalty parameters to find a solution that minimizes the objective function while satisfying the constraints.
  * The optimization path stored in all_positions can be used to visualize the algorithm's progress.

```python
# Augmented Lagrangian Method
def augmented_lagrangian_method(initial_x, max_outer_iter=20, max_inner_iter=100):
    x = initial_x
    lambda_eq = 0.0  # Lagrange multiplier for equality constraint
    lambda_ineq = 0.0  # Lagrange multiplier for inequality constraint
    rho_eq = 1.0  # Penalty parameter for equality constraint
    rho_ineq = 1.0  # Penalty parameter for inequality constraint

    all_positions = [x.copy()]  # Store all visited positions during optimization

    for outer_iter in range(max_outer_iter):
        # Perform gradient descent to minimize the augmented Lagrangian
        x, positions = minimize_augmented_lagrangian(x, lambda_eq, lambda_ineq, rho_eq, rho_ineq, max_inner_iter)
        all_positions.extend(positions)

        # Calculate constraint violations
        eq_violation = abs(equality_constraint(x))
        ineq_violation = max(0, inequality_constraint(x))

        # Print status
        print(f"Iteration {outer_iter}: x = {x}, eq_violation = {eq_violation}, ineq_violation = {ineq_violation}")

        # Check convergence
        if eq_violation < 1e-4 and ineq_violation < 1e-4:
            break

        # Update Lagrange multipliers and penalty parameters
        if eq_violation > 0.7 * rho_eq:
            lambda_eq += rho_eq * equality_constraint(x)
            rho_eq *= 2
        if ineq_violation > 0.7 * rho_ineq:
            lambda_ineq = max(0, lambda_ineq + rho_ineq * inequality_constraint(x))
            rho_ineq *= 2

    return np.array(all_positions)
```

```python
# Set up visualization
fig, ax = plt.subplots(figsize=(12, 10))

# Create meshgrid for contour plots
x1 = np.linspace(-2, 3, 200)
x2 = np.linspace(-2, 3, 200)
X1, X2 = np.meshgrid(x1, x2)
Z = np.zeros_like(X1)

# Calculate objective function values
for i in range(len(x1)):
    for j in range(len(x2)):
        Z[i, j] = objective_function(np.array([X1[i, j], X2[i, j]]))

# Plot objective function contours
contours = ax.contour(X1, X2, Z, levels=20, cmap='viridis')
plt.colorbar(contours, label='Objective Value')

# Plot equality constraint (circle)
theta = np.linspace(0, 2*np.pi, 100)
circle_x = np.cos(theta)
circle_y = np.sin(theta)
ax.plot(circle_x, circle_y, 'r--', label='Equality Constraint')

# Plot inequality constraint (line)
ax.plot([-2, 3], [2, -3], 'g--', label='Inequality Constraint')
ax.fill_between([-2, 3], [-2, -3], [3, 3], alpha=0.1, color='g')

# Run optimization
initial_point = np.array([0.0, 0.0])  # Starting point
positions = augmented_lagrangian_method(initial_point)

# Plot the optimization path
ax.plot(positions[:, 0], positions[:, 1], 'ro-', linewidth=1.5, markersize=3, label='Optimization Path')

# Final plot setup
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_title('Augmented Lagrangian Optimization with Constraints')
ax.grid(True)
ax.legend()
ax.set_aspect('equal')
plt.show()
```
